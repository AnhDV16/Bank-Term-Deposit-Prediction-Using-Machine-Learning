#### Change datatype

```python
def check_invalid(col, reg=r'^[0-9]+$', replace_reg=r'[0-9]'):
    # Ép kiểu chuỗi để tránh lỗi .str
    col = col.astype(str)

    invalid_row = col[~col.str.match(reg)]
    invalid_count = invalid_row.count()

    invalid = col.str.replace(replace_reg, '', regex=True).unique()

    return invalid, invalid_count
```

```python
for col in df[['age', 'balance', 'duration']].columns:
    print(col, check_invalid(df[col]))
```

```python
# Find 'NaN ' in all columns
df.map(lambda x: 'NaN' in str(x)).sum()
```

#### Converting Duration to Hours


```python

df['duration'] = df['duration'].apply(lambda n:n/60).round(2)
df
```

#### Removing Outliers

```python


sns.boxplot(data=df, x='duration')
plt.show()
```

```python
#Implementing the IQR Method to remove the outliers
q3 = df['duration'].quantile(.75)
q1 = df['duration'].quantile(.25)
iqr = q3-q1
iqr
```

```python
upperrange = q3+1.5*iqr
bottomrange = q1-1.5*iqr
df = df[(df['duration']>bottomrange) & (df['duration']<upperrange)]
df
```

```python
sns.boxplot(data=df,x='duration')
```

Goals of removing outliers in the duration column:
1. Increase higher correct analysis.

2. Cleansing data for ML (ML can be highly sensitive to outliers.)

3. Ensure ML correctly reflects typical behaviors of clients.
