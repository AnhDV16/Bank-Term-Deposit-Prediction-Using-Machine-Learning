```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the years 2022, 2023, and 2024
data = {
    'Chỉ tiêu': ['Tổng tiền gửi', 'Tiền gửi không kỳ hạn', 'Tiền gửi có kỳ hạn', 'Tiền gửi vốn chuyên dùng', 'Tiền gửi ký quỹ'],
    '2022 (%)': [100, 16.54, 82.26, 0.02, 1.17],
    '2023 (%)': [100, 17.03, 82.56, 0.02, 0.39],
    '2024 (%)': [100, 13.71, 85.85, 0.02, 0.42]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

# Create the bar chart for each year
ax.bar(df['Chỉ tiêu'], df['2022 (%)'], label='2022', alpha=0.6, width=0.2, align='center')
ax.bar(df['Chỉ tiêu'], df['2023 (%)'], label='2023', alpha=0.6, width=0.2, align='edge')
ax.bar(df['Chỉ tiêu'], df['2024 (%)'], label='2024', alpha=0.6, width=0.2, align='edge')

# Adding a line chart for each year
ax.plot(df['Chỉ tiêu'], df['2022 (%)'], marker='o', label='2022 (line)', color='b', linestyle='--')
ax.plot(df['Chỉ tiêu'], df['2023 (%)'], marker='o', label='2023 (line)', color='g', linestyle='--')
ax.plot(df['Chỉ tiêu'], df['2024 (%)'], marker='o', label='2024 (line)', color='r', linestyle='--')

# Labeling the chart
ax.set_title('Tỷ trọng Tiền Gửi theo các Năm (2022-2024)', fontsize=14)
ax.set_xlabel('Chỉ tiêu', fontsize=12)
ax.set_ylabel('Tỷ trọng (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
